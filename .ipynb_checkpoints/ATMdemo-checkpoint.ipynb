```python
Bal_Amount = 5000
PIN = int(input("Create 4-digit PIN"))
Amount = int(input("Enter Amount"))
pin = int(input("Enter your PIN"))
if pin == PIN:
    if Amount < Bal_Amount :
        print("""
        1. Cash Withdraw
        2. Bank transfer
        3. UPI Transaction
        """)
        choice = int(input("Enter Your Choice"))
        if choice == 1:
            print("Amount Withdrawn Successfully")
        if choice == 2:
            Acc_No = int(input("Enter The Acc_No"))
            IFSC = int(input("Enter The IFSC:"))
            print("Amount Transfered Successfully")
        if choice == 3:
            UPI = int(input("Enter The UPI:"))
            print("UPI Transaction Successfull")
        Bal_Amount -= Amount
        print("Remaining Amount Is: ", Bal_Amount)
    else :
        print("Insufficent Balence!")
else:
    print("Incorrect PIN")
```
