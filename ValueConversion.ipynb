```python
print("""----- Unit Converter -----
        1. Kilometers to Meters"
        2. Meters to Kilometers
        3. Celsius to Fahrenheit
        4. Fahrenheit to Celsius
        5. Kilograms to Chinese Catty (Jin)
        6. Chinese Catty (Jin) to Kilograms
        7. Feet to Meters
        8. Meters to Feet
        9. Exit
""")
while True:
    choice = input("\nEnter your choice (1-9): ")
    if choice == "9":
        print("Exiting Unit Converter... Bye!")
        break
    if choice not in ["1", "2", "3", "4", "5", "6", "7", "8"]:
        print("Invalid choice! Try again.")
        continue
    value = float(input("Enter value to convert: "))
    if choice == "1":
        print("Meters:", value * 1000)
    elif choice == "2":
        print("Kilometers:", value / 1000)
    elif choice == "3":
        print("Fahrenheit:", (value * 9/5) + 32)
    elif choice == "4":
        print("Celsius:", (value - 32) * 5/9)
    elif choice == "5":
        print("Chinese Catty (Jin):", value * 2)
    elif choice == "6":
        print("Kilograms:", value * 0.5)
    elif choice == "7":
        print("Meters:", value * 0.3048)
    elif choice == "8":
        print("Feet:", value / 0.3048)
```
