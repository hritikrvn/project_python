```python
import smtplib
import ssl 
from email.message import EmailMessage
EMAIL = "hritikshk01@gmail.com"
APP_PASSWORD = ""
RECEIVER = "kameshwarjamwal17@gmail.com"
def send_email(text):
    msg = EmailMessage()
    msg["From"] = EMAIL
    msg["To"] = RECEIVER
    msg["Subject"] = "ATM Transaction Alert"
    msg.set_content(text)
    context = ssl.create_default_context()
    with smtplib.SMTP_SSL("smtp.gmail.com", 465, context=context) as s:
        s.login(EMAIL, APP_PASSWORD)
        s.send_message(msg)
balance = 5000
pin = int(input("Create PIN: "))
print("ATM Started\n")
attempts = 3
while attempts > 0:
    enter = int(input("Enter PIN: "))
    if enter == pin:
        print("Login Successful\n")
        break
    else:
        attempts -= 1
        print("Wrong PIN. Attempts left:", attempts)
if attempts == 0:
    print("Card Blocked!")
    exit()
while True:
    print("""
    1. Check Balance
    2. Deposit
    3. Withdraw
    4. UPI Transfer
    5. Exit""")
    choice = input("Choose option: ")
    if choice == "1":
        print("Balance: ", balance)
    elif choice == "2":
        amount = float(input("Enter amount: "))
        balance += amt
        print("Deposited New Balance =", balance)
        send_email(f"Deposit: ₹{amount}\n Total Balance: ₹{balance}")
    elif choice == "3":
        amount = float(input("Enter amount for Withdrawal: "))
        if amount > balance:
            print("Not enough balance!")
        else:
            balance -= amount
            print("Withdraw Successful! Balance =", balance)
            send_email(f"Withdrawal: ₹{amount}\nRemaining Balance: ₹{balance}")
    elif choice == "4":
        amount = float(input("Enter amount: "))
        if amount > balance:
            print("Not enough balance!")
        else:
            upi = input("Enter UPI ID: ")
            balance -= amount
            print("Sent ₹", amount, "to", upi)
            send_email(f"UPI Transfer\nTo: {upi}\nAmount: ₹{amount}\nBalance: ₹{balance}")
    elif choice == "5":
        print("Thanks for using ATM!")
        break
    else:
        print("Invalid Choice!")
```
