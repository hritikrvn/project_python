```python
import time
from twilio.rest import Client
ACCOUNT_SID = " " 
AUTH_TOKEN = " "
TWILIO_NUMBER = "+19342457996"
RECEIVER_NUMBER = "+916206946290"
def send_sms(text):
    try:
        client = Client(ACCOUNT_SID, AUTH_TOKEN)
        message = client.messages.create(
            body=f"\n[ATM ALERT]\n{text}",
            from_=TWILIO_NUMBER,
            to=RECEIVER_NUMBER
        )
        print(">> SMS Sent Successfully!")
    except Exception as e:
        print(f">> Error sending SMS: {e}")
balance = 5000
pin = int(input("Create PIN: "))
print("\n--- ATM Started ---")
attempts = 3
while attempts > 0:
    enter = int(input("Enter PIN: "))
    if enter == pin:
        print("Login Successful\n")
        break
    else:
        attempts -= 1
        print("Wrong PIN. Attempts left :", attempts)
if attempts == 0:
    print("Card Blocked!")
    exit()
    print("""
    1. Check Balance
    2. Deposit
    3. Withdraw
    4. UPI Transfer
    5. Exit""")
    choice = input("Choose option: ")
    if choice == "1":
        print(f"Current Balance: ₹{balance}")
    elif choice == "2":
        amount = float(input("Enter amount to deposit: "))
        balance += amount  
        print(f"Deposited. New Balance = ₹{balance}")
        send_sms(f"Credited: ₹{amount}\nAvail Bal: ₹{balance}")
    elif choice == "3":
        amount = float(input("Enter amount for Withdrawal: "))
        if amount > balance:
            print("Not enough balance!")
        else:
            balance -= amount
            print(f"Withdraw Successful! Balance = ₹{balance}")
            send_sms(f"Debited: ₹{amount}\nAvail Bal: ₹{balance}")
    elif choice == "4":
        amount = float(input("Enter amount: "))
        if amount > balance:
            print("Not enough balance!")
        else:
            upi = input("Enter UPI ID: ")
            balance -= amount
            print(f"Sent ₹{amount} to {upi}")
            send_sms(f"UPI Transfer\nTo: {upi}\nAmt: ₹{amount}\nBal: ₹{balance}")
    elif choice == "5":
        print("Thanks for using ATM!")
        break
    else:
        print("Invalid Choice!"
```
